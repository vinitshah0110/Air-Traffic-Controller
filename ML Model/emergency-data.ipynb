```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../input/aqi-data/Airport_Data.csv')
data.head()
```

```python
df = pd.DataFrame(data[:20])
df.head()
```

```python
df['DIVERTED'] = df['DIVERTED'].apply(lambda item: np.random.randint(0,2))
```

```python
normal = []
emergency = []
      
for item in range(4):
    if df.iloc[item]['DIVERTED'] == 1:
        emergency.append([df.iloc[item]['FLIGHT_NUMBER'],df.iloc[item]['DESTINATION_AIRPORT']])
    else:
        normal.append([df.iloc[item]['FLIGHT_NUMBER'],df.iloc[item]['DESTINATION_AIRPORT']])
        
landing_queue = emergency + normal     
landing_queue
```
