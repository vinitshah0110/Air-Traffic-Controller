```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

```python
df = pd.read_csv('../input/aqi-data/Airport_Data.csv')
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report
from scipy.spatial.distance import cdist
```

```python
distortions = []
def check(airline,airport):
    for item in range(1,20):
        knn = KMeans(n_clusters=item)
        knn.fit(df[(df.AIRLINE==airline) & (df.ORIGIN_AIRPORT==airport) & (df.DESTINATION_AIRPORT!=airport)]
                                          [['LATITUDE','LONGITUDE']])
        distortions.append(sum(np.min(cdist(df[(df.AIRLINE==airline) & (df.ORIGIN_AIRPORT==airport) & (df.DESTINATION_AIRPORT!=airport)]
                                          [['LATITUDE','LONGITUDE']], knn.cluster_centers_,'euclidean'), axis=1)) / df[(df.AIRLINE==airline) & (df.ORIGIN_AIRPORT==airport) & (df.DESTINATION_AIRPORT!=airport)]
                                          [['LATITUDE','LONGITUDE']].shape[0])
check('VX','SFO')
```

```python
plt.plot(range(1,20), distortions, 'bx-');
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
```

```python
def cluster(airline,airport,desti_airport):
    km = KMeans(n_clusters=6)
    km.fit(df[(df.AIRLINE==airline) & (df.ORIGIN_AIRPORT==airport) & (df.DESTINATION_AIRPORT!=airport)]
                                          [['LATITUDE','LONGITUDE']])
    temp = pd.DataFrame(df[(df.AIRLINE==airline) & (df.ORIGIN_AIRPORT==airport) & (df.DESTINATION_AIRPORT!=airport)])
    temp['CLUSTER'] = km.labels_
    print(temp[(df.AIRLINE==airline) & (df.DESTINATION_AIRPORT==desti_airport)]['CLUSTER'].value_counts())
cluster('VX','SFO','LAX')
```
